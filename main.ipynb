## Cleaning

```python
# Handle NA values
df.isnull().sum()
```

```python
# drop all our missing values
df = df.dropna(axis=1)
```

## Picking X and y

```python
# Picking X and y

# Features (independent variables)
X = df[['MSSubClass', 'MSZoning', 'LotArea', 'Street', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'OverallQual','OverallCond', 'YearBuilt', 'YearRemodAdd', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'ExterQual', 'ExterCond', 'Foundation','BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'Heating','HeatingQC', 'CentralAir', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF','GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath','BedroomAbvGr', 'KitchenAbvGr', 'KitchenQual', 'TotRmsAbvGrd','Functional', 'Fireplaces', 'GarageCars', 'GarageArea', 'PavedDrive','WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch','ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold', 'YrSold', 'SaleType','SaleCondition']]

# Target (dependent variable)
y = df['SalePrice']
```

```python
# Train test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.3)
X_train.head()
```

## Normalization

### Numeric Values

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# Numeric Values

#select numerical columns for X_train
X_train_num = X_train.select_dtypes(include = ['number'])
X_train_num

# Fit and transform X_train
X_train_num_scaled = scaler.fit_transform(X_train_num)
X_train_num_scaled = pd.DataFrame(X_train_num_scaled, columns = X_train_num.columns)
X_train_num_scaled.head(2)
```

```python
# Select numerical columns for X_test
X_test_num = X_test.select_dtypes(include = ['number'])

# Transform numerical numerical X_test
X_test_num_scaled = scaler.transform(X_test_num)
X_test_num_scaled = pd.DataFrame(X_test_num_scaled, columns = X_test_num.columns)
X_test_num_scaled.head(2)
```

### Non-Numeric Values

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output = False, handle_unknown = "ignore")
```

```python
# Select categorical columns for X_train
X_train_cat = X_train.select_dtypes(include = [object])
X_train_cat

# Scale categorical columns for X_train
X_train_cat_scaled = ohe.fit_transform(X_train_cat)
X_train_cat_scaled = pd.DataFrame(X_train_cat_scaled, columns = ohe.get_feature_names_out())
X_train_cat_scaled.head(2)
```

```python
# Select categorical columns for X_test
X_test_cat = X_test.select_dtypes(include = [object])
X_test_cat

# Scale categorical columns for X_train
X_test_cat_scaled = ohe.transform(X_test_cat)
X_test_cat_scaled = pd.DataFrame(X_test_cat_scaled, columns = ohe.get_feature_names_out())
X_test_cat_scaled.head(2)
```

### combining (concatenating) the numeric and non-numeric normalized columns

```python
X_normalized_train = pd.concat((X_train_num_scaled, X_train_cat_scaled), axis =1)
X_normalized_train.head(2)
```

```python
X_normalized_test = pd.concat((X_test_num_scaled, X_test_cat_scaled), axis =1)
X_normalized_test.head(2)
```

### Try a Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()

lin_reg.fit(X_normalized_train, y_train)
```

### Calculate the MAE

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
y_pred = lin_reg.predict(X_normalized_test)
y_pred

# Mean absolute error
mean_abs_error = mean_absolute_error(y_test, y_pred)
print(f"MAE - { (mean_abs_error)}")

# Mean squared error
mean_sqd_error = mean_squared_error(y_test, y_pred)
print(f"MSE - { (mean_sqd_error)}")
```
